```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
seed = 0
df1 = pd.read_csv("results/high_dim_11-12-22_seed_{}_gradient_type_partial_deriv_loss_beta_train.csv".format(seed))
df2 = pd.read_csv("results/high_dim_11-12-22_seed_{}_gradient_type_total_deriv_train.csv".format(seed))
df3 = pd.read_csv("results/high_dim_11-12-22_seed_{}_gradient_type_expected_gradient_beta_naive_train.csv".format(seed))
```

```python
def conversion_method(arr):
    l = []
    for s in arr:
        t = s[1:-1].split(", ")
        t = [float(x) for x in t]
        l.append(np.array(t).reshape(10, 1))
    return l
```

```python
plt.figure(figsize=(16/3, 4))
plt.plot(df1["iteration"], -df1["loss"], label="Model Effect")
plt.plot(df2["iteration"], -df2["loss"], label="Policy Effect")
plt.plot(df3["iteration"], -df3["loss"], label="Naive Gradient")
plt.xlabel("Iteration", fontsize=15)
plt.ylabel("Equilibrium Policy Value  " + r"$\hat{V}_{eq}(\hat{\beta})$", fontsize=15)
plt.legend(fontsize=12)
plt.title("High-Dimensional Synthetic \n Equilibrium Policy Value vs. Iteration", fontsize=15)
plt.savefig("figs/paper-figure-4.pdf", bbox_inches='tight')
```
