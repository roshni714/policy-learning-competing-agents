```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv("results/nels_etas_12-11-23_b_seed_0_method_total_deriv_train.csv")
df1 = pd.read_csv("results/nels_months_attended_12-11-23_b_seed_0_method_total_deriv_train.csv")
df3 = pd.read_csv("results/nels_months_attended_12-11-23_b_seed_0_method_total_deriv_train.csv")

dfs = [df1, df2, df3]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(24, 6))
for i, df in enumerate(dfs):
    ax[i].plot(dfs[i]["iteration"], dfs[i]["mag_mg"], label=r"$|\tau_{MG}|$")
    ax[i].plot(dfs[i]["iteration"], dfs[i]["mag_eg"], label=r"$|\tau_{EG}|$")
    ax[i].plot(dfs[i]["iteration"], dfs[i]["mag_pg"], label=r"$|\tau_{PG}|$")
    ax[i].set_xlabel("Iteration", fontsize=15)
    ax[i].set_ylabel("Magnitude", fontsize=15)
    ax[i].legend()
ax[0].set_title("Magnitude of Gradient Components of " + r"$\hat{V}_{eq, 1}(\beta)$", fontsize=15)
ax[1].set_title("Magnitude of Gradient Components of " + r"$\hat{V}_{eq, 2}(\beta)$", fontsize=15)
ax[2].set_title("Magnitude of Gradient Components of " + r"$\hat{V}_{eq, 3}(\beta)$", fontsize=15)

plt.savefig("figs/paper-figure-10.pdf")
```
